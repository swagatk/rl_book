```python
import pandas as pd
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=1)

df1 = pd.read_csv('taxi_qlearn.tsv', delimiter='\t')
df2 = pd.read_csv('taxi_sarsa.tsv', delimiter='\t')
df1.columns = ['avg_ep_reward', 'epsilon']
df2.columns = ['avg_ep_reward', 'epsilon']
df1.plot(y='avg_ep_reward', label='qlearn', ax=axes[0])
df2.plot(y='avg_ep_reward', label='sarsa', ax=axes[0])
df1.plot(y='epsilon', label='qlearn', ax=axes[1])
df2.plot(y='epsilon', label='sarsa', ax=axes[1])
axes[0].grid(linewidth=0.5,color='gray')
axes[1].grid(linewidth=0.5,color='gray')
axes[1].set_xlabel('Training Episodes')
axes[0].set_ylabel('Average Episodic Reward')
axes[1].set_ylabel('Epsilon')
plt.show()
```
